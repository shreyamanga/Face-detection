```python
import cv2
```

```python
face = cv2.imread("face_photo.jpg")
face
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.imshow(face)
plt.show()
```

```python
face1 = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
plt.imshow(face1)
```

```python
haar_cascade = cv2.CascadeClassifier('FaceCascade\FaceCascade\haarcascade_frontalface_alt.xml')
haar_cascade
```

```python
cv2.imshow('second image',face1)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
face_det = haar_cascade.detectMultiScale(face1, 1.2, 3)
```

```python
def convertToRGB(face1):
    return cv2.cvtColor(face1, cv2.COLOR_GRAY2RGB)
for (x,y,w,h) in face_det:
     cv2.rectangle(face, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(convertToRGB(face1))
```

```python
import cv2
cap = cv2.VideoCapture(0)
while(True):
    ret, imge = cap.read()
    gray = cv2.cvtColor(imge, cv2.COLOR_BGR2GRAY)
    face_det = haar_cascade.detectMultiScale(gray, 1.2, 5)
    for (x,y,w,h) in face_det:
        cv2.rectangle(gray, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.imshow('face',imge)
    k = cv2.waitKey(10) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
